```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('population_data.csv')
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# choosing top 20 countries for visualization
a=df.head(20)
a
```

```python
plt.figure(figsize=(16,8))
sns.barplot(a['Country Name'],a['2022'],data=a,palette='coolwarm')
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Population in 2022', fontsize=12)
plt.title("Countrywise Population in 2022", fontsize=16)
plt.xticks(rotation= 45)
plt.show()
```

```python
data = df.groupby('Country Name')['1960','2020','2021','2022'].sum().sort_values(by="2022",ascending=False).head(50).reset_index()
data.head()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data['Country Name'],data['2022'],data=data,palette='cool')
plt.xlabel('Territory/Country', fontsize=12)
plt.ylabel('Population in 2022', fontsize=12)
plt.title("Top 50 Populated Territory in 2022", fontsize=16)
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data['Country Name'],data['1960'],data=data,palette='rainbow')
plt.xlabel('Territory/Country', fontsize=12)
plt.ylabel('Population in 1960', fontsize=12)
plt.title("Top 50 Populated Territory in 1960", fontsize=16)
plt.xticks(rotation=90)
plt.show()
```
